# Downloading and Running An LLM

Load the Phi-3-mini-4k-instruct model onto the GPU, same as Chapter 1.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

Unlike chapter 1, this time _don't_ load tranformers.pipline to encapsulate the model, tokenizer, and text-generation process into a single pipeline function. This way, we can explore them separately.

In the code cell below, the prompt is converted to tokens, the model processes the tokens and returns its generative result as tokens, and those tokens are converted back to readable text: **Subject: Sincere Apologies for the Gardening Mishap... Dear**.

```python
prompt = "Write an email apologizing to Sarah for the tragic gardening mishap. Explain how it happened.<|assistant|>"

# Tokenize the input prompt and return the tokens as PyTorch (pt) tensors.
# `.input_ids.to()` converts the tokens to a CUDA tensor so they can be
# processed on the GPU.
input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")

# Generate text from the prompt tokens, stopping at 20 tokens of response.
generation_output = model.generate(
  input_ids=input_ids,
  max_new_tokens=20
)

# Decode the response tokens back into readable text.
print(tokenizer.decode(generation_output[0]))
```

`input_ids` is a tensor of index values for reference table inside the tokenizer.

```python
print(input_ids)
```

Here's how it tokenized the prompt. Use the `tokenizer.decode()` method on the prompt tokens. Notice the first token is "\<s\>". That's a special indicator token identifying the beginning of text. Tokens are not always words or even stems. Punctuation marks are their own tokens.

```python
for id in input_ids[0]:
   print(tokenizer.decode(id))
```

So how does a tokenizer tokenize text? There are several tokenization methods. Byte pair encoding (BPE, used in many GPT models), and WordPiece (used by BERT) are two. After selecting vocabulary size and other settings, the tokenizer is trainined on a corpus of text to establish the best vocabulary.

Tokenization could be done at the word level. Word2vec does word tokenization. But this results in many tokens with minimal differences (e.g., apology, apologize, apologetic, apologist), and the increased potential for encountering new words. At the other extreme, you could tokenize by character, but that turns out to be difficult to model. There is another method to be aware of called _byte tokens_ that represent unicode characters.

# Comparing Trained LLM Tokenizers

Tokenizers differ in their tokenization method, parameters and specialization tokens used to initialize, and their training dataset. Below we define a function that tokenizes a `sentence` using the requested `tokenizer_name`. The `colors_list` helps visualize the different tokens.

```python
# from transformers import AutoModelForCausalLM, AutoTokenizer

colors_list = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def show_tokens(sentence, tokenizer_name):
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    token_ids = tokenizer(sentence).input_ids
    for idx, t in enumerate(token_ids):
        print(
            f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m' +
            tokenizer.decode(t) +
            '\x1b[0m',
            end=' '
        )
```

We'll use the following text to show how tokenizers treat capitalization, non-English language, emojis, programming code, and numbers.

```python
text = """
English and CAPITALIZATION
ðŸŽµ é¸Ÿ
show_tokens False None elif == >= else: two tabs:"    " Three tabs: "       "
12.0*50=600
"""
```

First up is **BERT base model (uncased)** ([Hugging Face](https://huggingface.co/google-bert/bert-base-uncased), 2018). It uses WordPiece to tokenize. Its vocabular is 30,522. Unknown tokens show up as \[UNK\]. The emoji and Chinese chars are \[UNK\]. Newline breaks are gone - the model is blind to information implied by newlines.

```python
show_tokens(text, "bert-base-uncased")
```

# Contextualized Word Embeddings From a Language Model (Like BERT)

After the tokenizer is initialized and trained, it is used to train the LLM. This is why LLMs and tokenizers are always linked. The LLM has an embedding vector for each token in the tokenizer vocabulary. Initially the embedding vectors are initialized to something random, then the training process optimizes the values.

Instead of representing each token with a static vector, LLMs created contextualized word embeddings. Let's see how this is done using the DeBERTa v3 LLM. `tokenizer()` tokenizes "Hello world", and `model()` embeds each token into a 384-value vector.

```python
from transformers import AutoModel, AutoTokenizer

# Load a tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-base")

# Load a language model
model = AutoModel.from_pretrained("microsoft/deberta-v3-xsmall")

# Tokenize the sentence
tokens = tokenizer('Hello world', return_tensors='pt')

# Process the tokens
output = model(**tokens)[0]
```

`tokens` is a multidimensional array. The first dimension is the batch dimension used in cases (like training). The other dimensions are 4 tokens, each embedded in a 384-value vector.

```python
output.shape
```

Here are the four tokens, translated back into English.

```python
for token in tokens['input_ids'][0]:
    print(tokenizer.decode(token))
```

And here are the four vectors.

```python
output
```

# Text Embeddings (For Sentences and Whole Documents)

LLMs usually operate on entire sentences, paragraphs, or whole documents. Special models produce _text embeddings_, a single vector the represents pieces of text. One way to produce a text embedding is to average over the token embeddings. However, high quality text embedding models have been developed. Use the sentence-transformers package to load publicly available text embedding models. Let's try this with all-mpnet-base-v2 ([Hugging Face](https://huggingface.co/sentence-transformers/all-mpnet-base-v2)). It maps sentences and paragraphs to a 768 dimensional dense vector space.

```python
from sentence_transformers import SentenceTransformer

# Load model
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')

# Convert text to text embeddings
vector = model.encode("Best movie ever!")
```

 The sentence is now enoded into a 768-value vector.

```python
vector.shape
```

Like LLMs, word2vec is trained on text examples. The algorithm uses a sliding window to generate training examples. The embeddings are generated from a classification task using a neural network to predict whether words commonly appear in the same context. For example, given the phrase "Thou shalt not make a machine in the likeness of a human mind", the algorith might form a sliding window 2 to the left and right of each word, prediting "1" for "Thou", "shalt", "Make", and "a" for the middle word "not". To create negative examples, the text would include negative examples too - random words labeled "0". After creating millions (or even billions) of examples, an randomly initialized embedding vector is created for each token, a matrix of vocab-size x embedding-dimension-size. The model trains to predict if two embedding vectors are related, updates the embeddings, and repeats.

# Recommending songs by embeddings

Let's use the word2vec algorithm to embed songs using human-made music playlists. Think of each song as a token, and the playlist as a sentence. We'll use a dataset of 100s of radio station playlists. Start by loading the dataset.

```python
import pandas as pd
from urllib import request

# Get the playlist dataset file
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')

# Parse the playlist dataset file. Skip the first two lines as
# they only contain metadata
lines = data.read().decode("utf-8").split('\n')[2:]

# Remove playlists with only one song
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]

# Load song metadata
songs_file = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt')
songs_file = songs_file.read().decode("utf-8").split('\n')
songs = [s.rstrip().split('\t') for s in songs_file]
songs_df = pd.DataFrame(data=songs, columns = ['id', 'title', 'artist'])
songs_df = songs_df.set_index('id')
```

Each element of the playlist is a list of song IDs.

```python
print( 'Playlist #1:\n ', playlists[0], '\n')
print( 'Playlist #2:\n ', playlists[1])
```

Train the model.

```python
from gensim.models import Word2Vec

# Train our Word2Vec model
model = Word2Vec(
    playlists, vector_size=32, window=20, negative=50, min_count=1, workers=4
)
```

Use the embeddings to find similar songs. We'll work with "Fade to Black", song id 2172.

```python
song_id = 2172

print(songs_df.iloc[song_id])
```

Here are the most similar songs.

```python
# Ask the model for songs similar to song #2172
model.wv.most_similar(positive=str(song_id))
```

```python
print(songs_df.iloc[2172])
```

Let's get their names.

```python
import numpy as np

def print_recommendations(song_id):
    similar_songs = np.array(
        model.wv.most_similar(positive=str(song_id),topn=5)
    )[:,0]
    return  songs_df.iloc[similar_songs]

# Extract recommendations
print_recommendations(2172)
```
